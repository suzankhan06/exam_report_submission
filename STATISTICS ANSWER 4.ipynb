```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
y = pd.Series([1, 2, 3, 4, 3, 5, 4])
x = pd.Series([1, 2, 3, 4, 5, 6, 7])
correlation = y.corr(x)
correlation 
```

```python
# adds the title
plt.title('Correlation')
 
# plot the data
plt.scatter(x, y)
 
# fits the best fitting line to the data
plt.plot(np.unique(x), 
         np.poly1d(np.polyfit(x, y, 1))
         (np.unique(x)), color='red')
 
# Labelling axes
plt.xlabel('x axis')
plt.ylabel('y axis')
```

Insight: here we can see that our x variable is positively co related with y varibale
